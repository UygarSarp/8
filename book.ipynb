```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

# How many chapters

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]+")
findings = re.findall(pattern, book)
len(findings)
```

# The sentences that haves love in it

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^             ]*.")
findings = re.findall(pattern, book)
len(findings)
```

# Most common words

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book)
d = {}
for i in findings:
    i = i.lower()
    if i in d.keys():
        d[i] = d[i] + 1
    else:
        d[i] = 1
d_list = [(value, key) for (key, value) in d.items() ]
d_list = sorted(d_list, reverse=True)
```

# paragraphs that used love

```python
pattern = re.compile("[^\n]*[^a-zA-Z]+love[^a-zA-Z]+[^\n]*.")
findings = re.findall(pattern, book)
findings
```

# Chapter Titles

```python
pattern = re.compile("([a-zA-Z ]+)\n\n")
findings = re.findall(pattern, book)
findings
```

# Function finds words

```python
def find(word):
    pattern = re.compile(f"{word}[^a-zA-Z]+")
    findings = re.findall(pattern, book)
    if len(findings) == 0:
        print(f"{word} does not exist in the book.")
    else:
        return len(findings)
```

```python
find("love")
```

```python
find("hate")
```
